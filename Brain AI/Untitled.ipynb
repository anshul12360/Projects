```python
import pyttsx3
import datetime
import speech_recognition as sr
import wikipedia
import webbrowser
import random
import os
```

```python
engine = pyttsx3.init('sapi5')
voices = engine.getProperty('voices')
#print(voices[1].id)
engine.setProperty('voice',voices[0].id)
```

```python
def speak(audio):
    engine.say(audio)
    engine.runAndWait()
```

```python
def wishme():
    hour = int(datetime.datetime.now().hour)
    if hour>=0 and hour<12:
        speak("Good Morning Anshul Sir !")
    elif hour>=12 and hour<18:
        speak("Good Afternoon Anshul Sir !")
    else:
        speak("Good Evening Anshul Sir!")
        
    speak("I am BRAIN-A  I . Please tell me how may I help you")
    
    
```

```python
def takeCommand():
    r = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening")
        r.pause_threshold=1
        audio = r.listen(source)
    try:
        print("Recognizing....")
        query = r.recognize_google(audio, language='en-in')
        print(f"User said: {query}\n",)
    except Exception as e:
        print("Say that again please....")
        return "None"
    return query
```
